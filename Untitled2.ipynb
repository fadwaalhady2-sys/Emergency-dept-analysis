```python
import pandas as pd
df= pd.read_csv('edstays.csv.gz',compression='gzip')
print(df.head())
print(df.columns.tolist())
```

```python
df['intime'] = pd.to_datetime(df['intime'])
df['outtime'] = pd.to_datetime(df['outtime'])

# حساب مدة البقاء لكل مريض (بالساعات)
df['duration_hours'] = (df['outtime'] - df['intime']).dt.total_seconds() / 3600

# طباعة متوسط مدة البقاء
print(f"متوسط مدة بقاء المريض في الطوارئ: {df['duration_hours'].mean():.2f} ساعة")
```

```python

df_diag = pd.read_csv('diagnosis.csv.gz', compression='gzip')

# ندمج بناءً على رقم الزيارة (stay_id)
merged_df = pd.merge(df, df_diag, on='stay_id')

# 3. حساب متوسط مدة البقاء لكل تشخيص
# نختار أهم 10 تشخيصات من حيث التكرار ليكون التحليل دقيقاً
top_diagnoses = merged_df.groupby('icd_title')['duration_hours'].mean().sort_values(ascending=False).head(10)

print("أكثر 10 تشخيصات استغراقاً للوقت في الطوارئ (بالساعات):")
print(top_diagnoses)
```

```python
import matplotlib.pyplot as plt

# رسم بياني لأكثر 10 تشخيصات استغراقاً للوقت
plt.figure(figsize=(10, 6))
top_diagnoses.plot(kind='bar', color='skyblue')

plt.title('Top 10 Diagnoses by Duration in ED')
plt.xlabel('Diagnosis')
plt.ylabel('Average Duration (Hours)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# عرض الرسم
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. اختصار الأسماء الطويلة (نأخذ أول 20 حرفاً فقط)
shortened_names = [name[:20] + "..." if len(name) > 20 else name for name in top_diagnoses.index]

# 2. تحسين الرسم (تغيير اللون، الخطوط، وتنسيق النص)
plt.figure(figsize=(12, 7))
bars = plt.bar(shortened_names, top_diagnoses.values, color='#4A90E2', edgecolor='white')

# 3. تحسين العناوين باللغة الإنجليزية
plt.title('Top 10 Diagnoses by Duration in ED', fontsize=16, pad=20)
plt.xlabel('Diagnosis (Shortened)', fontsize=12)
plt.ylabel('Average Duration (Hours)', fontsize=12)

# 4. تدوير الأسماء وتحسين التنسيق
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
df['arrival_hour'] = df['intime'].dt.hour

# حساب متوسط مدة البقاء لكل ساعة
hourly_trends = df.groupby('arrival_hour')['duration_hours'].mean()

# رسم بياني للاتجاهات الزمنية
hourly_trends.plot(kind='line', marker='o', figsize=(10, 5))
plt.title('Patient Load vs. Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Stay (Hours)')
plt.show()
```

```python
df_vitals = pd.read_csv('vitalsign.csv.gz', compression='gzip')

# 2. دمجها مع جدولنا الرئيسي (df) عبر stay_id
merged_vitals = pd.merge(df, df_vitals, on='stay_id')

# 3. تحليل سريع: متوسط نبض القلب (heartrate) حسب حالة المريض
# هذا يعطي دلالة طبية قوية جداً
vitals_impact = merged_vitals.groupby('disposition')['heartrate'].mean()
print("متوسط نبض القلب حسب حالة الخروج:")
print(vitals_impact)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
# رسم Box Plot يوضح توزيع نبض القلب لكل حالة خروج
sns.boxplot(x='disposition', y='heartrate', data=merged_vitals, palette='viridis')

plt.title('Distribution of Heart Rate by Patient Disposition', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. اختيار المتغيرات التي نريد دراسة ارتباطها
# سنحتاج هنا لإضافة العمر 'age' إذا كان متاحاً في بياناتك، أو نكتفي بما لدينا
subset = merged_vitals[['duration_hours', 'heartrate']]

# 2. حساب مصفوفة الارتباط (Correlation Matrix)
corr_matrix = subset.corr()

# 3. رسم الخريطة الحرارية
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap: Duration vs Heart Rate', fontsize=14)
plt.show()
```

```python

missing_data = merged_vitals.isnull().sum()

# 2. حساب النسبة المئوية للبيانات المفقودة
missing_percentage = (merged_vitals.isnull().sum() / len(merged_vitals)) * 100

# 3. عرض النتائج في جدول مرتب
data_quality_report = pd.DataFrame({'Missing Values': missing_data, 'Percentage (%)': missing_percentage})
print("تقرير جودة البيانات (القيم المفقودة):")
print(data_quality_report[data_quality_report['Missing Values'] > 0].sort_values(by='Percentage (%)', ascending=False))
```
